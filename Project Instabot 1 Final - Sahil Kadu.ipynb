# Question 6.2

```python
#first click on my profile
my_prof = driver.find_element_by_class_name('_aaav')
my_prof.click()

#click my profile from dropdown
link = driver.find_element_by_xpath('//*[@id="mount_0_0_in"]/div/div[1]/div/div[1]/div/div/div[1]/div[1]/section/nav/div[2]/div/div/div[3]/div/div[6]/div[2]/div[2]/div[2]/a[1]')
link.click()

#click on followers
followers = WebDriverWait(driver, 10).until(EC.element_to_be_clickable((By.XPATH, '//a[@href = "/username/followers/"]')))
followers.click()
sleep(3)

#create set to add all my followers
pop_up_2 = driver.find_element_by_xpath('//*[@id="mount_0_0_in"]/div/div[1]/div/div[2]/div/div/div[1]/div/div[2]/div/div/div/div/div/div/div/div[2]')
count = 0
my_foll = set()
while count<250:
    a = driver.find_elements_by_xpath('//*[@id="mount_0_0_in"]/div/div[1]/div/div[2]/div/div/div[1]/div/div[2]/div/div/div/div/div/div/div/div[2]/ul/div/li')
    for i in range(len(a)):
        my_foll.add(a[i].text.split('\n')[0])
        count+=1   
    driver.execute_script('arguments[0].scrollTop = arguments[0].scrollTop + arguments[0].offsetHeight;', pop_up_2)
    sleep(3)

#close followers dialog box
close = driver.find_element_by_xpath('//*[@id="mount_0_0_in"]/div/div[1]/div/div[2]/div/div/div[1]/div/div[2]/div/div/div/div/div/div/div/div[1]/div/div[3]')
close.click()

#search foodtalkindia
search = driver.find_element_by_xpath('//input[@placeholder = "Search"]')
search.send_keys('foodtalkindia')
sleep(2) 
foodtalkindia = driver.find_element_by_xpath('//a[@href = "/foodtalkindia/"]')
foodtalkindia.click()
sleep(3)

#search mutual followers & create set for it
m = driver.find_elements_by_class_name('_aaai')
mutual_followers = []
for i in range(len(m)):
    mutual_followers.append(m[i].text)

mu_foll = set()
for i in range(1, len(mutual_followers)):
    mu_foll.add(mutual_followers[i])
mu_foll

#find the difference between two sets so that we get followers of foodtalkindia who do not follow me
for ele in mu_foll.difference(my_foll):
    print(ele)


```
